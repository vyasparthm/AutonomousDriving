## Finding Corners

This notebook will explain the very first step to undistort images.
we will use Chess board as an example, why? because we all know how the undistored chess board looks like and its a great high contras image.

We will use the OpenCV functions `findChessboardCorners()` and `drawChessboardCorners()` to automatically find and draw corners in an image of a chessboard pattern.

For official documentation of these two visit OpenCV docs: 
[cv2.findChessboardCorners()](https://docs.opencv.org/2.4/modules/calib3d/doc/camera_calibration_and_3d_reconstruction.html#cv2.findChessboardCorners) and [cv2.drawChessboardCorners()](https://docs.opencv.org/2.4/modules/calib3d/doc/camera_calibration_and_3d_reconstruction.html#cv2.drawChessboardCorners).

By applying these functions, your image shall look like this: <br/>

<img src="./img/foundcorners.jpg"/>

```python
import matplotlib
import cv2
import numpy

# Read the image
input_img = './img/calibration_test.png'
img = cv2.imread(input_img)

# Convert image to Grayscale
gray = cv2.cvtColor(img,cv2.COLOR_RGB2GRAY)

# Feed Gray image to findChessboardCorners()
ret, corners = cv2.findChessboardCorners(gray,(8,6),None)
#print(ret)

# Lets draw drawChessboardCorners

img = cv2.drawChessboardCorners(img,(8,6),corners, ret)

# Show if pattern was found

if ret == True:
    plt.imshow(img)
    plt.imsave('./img/foundcorners.jpg',img)

```
