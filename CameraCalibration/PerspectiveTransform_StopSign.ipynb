## Perspective Transform on a Stop Sign

In **Perspective Transformation**, , we can change the perspective of a given image or video for getting better insights about the required information. In Perspective Transformation, we need provide the points on the image from which want to gather information by changing the perspective. We also need to provide the points inside which we want to display our image. Then, we get the perspective transform from the two given set of points and wrap it with the original image.

We use `cv2.getPerspectiveTransform` and then `cv2.wrapPerspective` .

We will do this by a transforming an image with STOP sign.

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
import matplotlib.image as mpimg


# Read original/source image
img = mpimg.imread('./img/STOPSign.jpg')

plt.imshow(img)
```

```python
%matplotlib qt
plt.imshow(img)
plt.plot(642,242,'.') #Top Right
plt.plot(659,339,'.') #bottom Right
plt.plot(410,265,'.') #Bottom Left
plt.plot(412,179,'.') #Top Left

waitKey(20)
```

If the above image points are acceptable then lets write a function to to transform the perspective

```python
def warp(img):
    """
    Takes an image as an input and produces a transformed image
    """
    
    # Define calibration box in source and destination coordinates
    img_size = (img.shape[1],img.shape[0])
    
    #four source coordinates
    src = np.float32(
    [[642,242],
    [659,339],
    [410,265],
    [412,179]])
    
    #four Destination coordinates
    dst = np.float32(
    [[660,240],
    [660,270],
    [410,270],
    [410,240]])
    
    # Get perspective transform Matrix
    M = cv2.getPerspectiveTransform(src,dst)
    
    # Create Warped Image - It uses Linear interpolation
    warped = cv2.warpPerspective(img,M,img_size, flags= cv2.INTER_LINEAR)
    
    return warped
```

```python
%matplotlib inline
warped_im = warp(img)

f,(ax1,ax2) = plt.subplots(1,2,figsize=(20,20))

ax1.set_title('Source')
ax1.imshow(img)
ax2.set_title('Transformed')
ax2.imshow(warped_im)

```

Although this doesn't look like its a great transform but its close to what a person sees standing in front.
